非正規化クラスタリング（伸びのみ正規化：各クラスタのレンジをそろえるため）


```python
import numpy as np
import pandas as pd
import os
from scipy.interpolate import interp1d
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def preprocess_waveform_non_normalized(df, target_length=100):
    x = df['Nominal strain'].values
    y = df['Nominal stress'].values

    if len(x) < 2 or len(y) < 2:
        raise ValueError("補間に必要なデータポイントが不足しています。")

    x_unique, x_indices = np.unique(x, return_index=True)
    y_unique = y[x_indices]

    try:
        x_segment = x_unique
        y_segment = y_unique

        f = interp1d(x_segment, y_segment, kind='linear', fill_value="extrapolate", bounds_error=False)
        x_new = np.linspace(min(x_segment), max(x_segment), target_length)
        y_new = f(x_new)

    except Exception as e:
        print(f"波形データの処理中にエラーが発生しました: {e}")
        y_new = np.zeros(target_length)

    return y_new, np.max(y_unique), x_unique[np.argmax(y_unique)]

def load_and_process_data_non_normalized(data_folder, target_length=100):
    processed_data = []
    ts_values = []
    εts_values = []
    filenames = []
    for filename in os.listdir(data_folder):
        if filename.endswith('.csv'):
            filepath = os.path.join(data_folder, filename)
            df = pd.read_csv(filepath)
            if 'Nominal strain' not in df.columns or 'Nominal stress' not in df.columns:
                raise KeyError(f"{filename}に'Nominal strain'または'Nominal stress'が見つかりません")
            df = df.dropna()
            processed_wave, TS, εTS = preprocess_waveform_non_normalized(df, target_length)
            processed_data.append(processed_wave)
            ts_values.append(TS)
            εts_values.append(εTS)
            filenames.append(filename)
    return np.array(processed_data), np.array(ts_values), np.array(εts_values), filenames

def plot_clusters_non_normalized(data, labels, centers, filenames):
    n_clusters = len(np.unique(labels))
    fig, axes = plt.subplots(2, (n_clusters + 1) // 2, figsize=(30, 16))
    axes = axes.ravel()
    colors = plt.cm.viridis(np.linspace(0, 1, n_clusters))

    title_font_size = 32
    label_font_size = 30
    ticks_font_size = 24
    clearance_size = 20

    cluster_files = [[] for _ in range(n_clusters)]  # 各クラスタごとのファイル名を格納するリスト

    for i in range(n_clusters):
        cluster_data = data[labels == i]
        for wave in cluster_data:
            axes[i].plot(np.linspace(0, 1, len(wave)), wave, alpha=1, color=colors[i])

        axes[i].plot(np.linspace(0, 1, len(centers[i])), centers[i], linewidth=6, color='black', linestyle='--')

        # クラスタ内のファイル名をリストアップ
        cluster_indices = np.where(labels == i)[0]
        for index in cluster_indices:
            cluster_files[i].append(filenames[index])

        axes[i].set_title(f'Cluster {i}', fontsize=title_font_size, pad=clearance_size)
        axes[i].set_xlabel('Nominal Strain', fontsize=label_font_size, labelpad=clearance_size)
        axes[i].set_ylabel('Nominal Stress', fontsize=label_font_size, labelpad=clearance_size)
        axes[i].tick_params(axis='both', labelsize=ticks_font_size)

    # 各クラスタのファイル名を印刷
    for i, files in enumerate(cluster_files):
        print(f'Cluster {i} contains the following files: {files}')

    # 代表的な凡例を作成
    handles = [
        plt.Line2D([0], [0], color='black', linewidth=2, label='Input Data (Line)'),
        plt.Line2D([0], [0], color='black', linestyle='--', linewidth=4, label='Cluster Center (Dashed)')
    ]

    # プロット内の凡例設定（2段）
    plt.legend(handles=handles, fontsize=25, loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)

    plt.tight_layout()
    plt.show()

def main_clustering_non_normalized():
    data_folder = '/content/drive/MyDrive/program_LSTM/output-data_forecast_LSTM'
    target_length = 100
    processed_data, ts_values, εts_values, filenames = load_and_process_data_non_normalized(data_folder, target_length)

    n_clusters = 4
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(processed_data)
    centers = kmeans.cluster_centers_

    plot_clusters_non_normalized(processed_data, labels, centers, filenames)

    return centers, ts_values, εts_values, labels, filenames

# Execute
centers, ts_values, εts_values, labels, filenames = main_clustering_non_normalized()
```

PCA：直前のクラスタリング結果に準拠
*   標準化しないのがポイント
*   既にクラスタリングで揃えらているので
*   クラスタリングせずにPCAから始める場合は正規化必要（Nominal strain）






```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import pickle

# 前のセルからcenters, ts_values, labelsを受け取る
# ts_centersの計算
ts_centers = np.array([np.mean(ts_values[labels == i]) if any(labels == i) else np.nan for i in range(len(centers))])
valid_indices = ~np.isnan(ts_centers)
ts_centers = ts_centers[valid_indices]
centers = centers[valid_indices]

def apply_pca(centers):
    # データの標準化（スケーリング）
    #scaler = StandardScaler()
    #scaled_centers = scaler.fit_transform(centers)

    pca = PCA(n_components=1)
    pca_transformed_centers = pca.fit_transform(centers)
    return pca_transformed_centers, pca

def perform_polynomial_regression(ts_values, pca_coord, degree=2):
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])
    model.fit(ts_values.reshape(-1, 1), pca_coord)
    return model

def reverse_pca_check(pca_model, pca_coords, original_data):
    reconstructed_data = pca_model.inverse_transform(pca_coords)
    for i, (original, reconstructed) in enumerate(zip(original_data, reconstructed_data)):
        print(f"Original (first 5): {original[:5]}")
        print(f"Reconstructed (first 5): {reconstructed[:5]}")
        print(f"Difference: {np.linalg.norm(original - reconstructed)}")
        print("=" * 50)

# PCAの適用
pca_coords, pca_model = apply_pca(centers)

# PCAの可逆性チェック
reverse_pca_check(pca_model, pca_coords, centers)

# スケーリングの準備
ts_scaler = StandardScaler()
ts_centers_scaled = ts_scaler.fit_transform(ts_centers.reshape(-1, 1))

# PCAスケーリング用
pca_scaler = StandardScaler()
pca_coords_scaled = pca_scaler.fit_transform(pca_coords)

# 回帰の実行
model_pca1_poly = perform_polynomial_regression(ts_centers_scaled.flatten(), pca_coords_scaled.flatten(), degree=2)

# モデルとスケーラーの保存
with open('pca_model.pkl', 'wb') as f:
    pickle.dump(pca_model, f)
with open('scaler_ts.pkl', 'wb') as f:
    pickle.dump(ts_scaler, f)
with open('scaler_pca.pkl', 'wb') as f:
    pickle.dump(pca_scaler, f)
with open('model_pca1_poly.pkl', 'wb') as f:
    pickle.dump(model_pca1_poly, f)

# 回帰式の表示
coefs = model_pca1_poly.named_steps['linear'].coef_
intercept = model_pca1_poly.named_steps['linear'].intercept_
print(f"PCA1 Regression Equation: PCA1 = {coefs[2]} * TS^2 + {coefs[1]} * TS + {intercept}")

# 可視化
title_font_size = 32
label_font_size = 30
legend_font_size = 27
ticks_font_size = 24
clearance_size=20

plt.figure(figsize=(8, 6))
plt.scatter(ts_centers, pca_coords, label='PCA Component 1')
plt.xlabel('Average TS Value', fontsize=label_font_size, labelpad=clearance_size)
plt.ylabel('PCA Component 1', fontsize=label_font_size, labelpad=clearance_size)
plt.title('PCA of Cluster Centers with Average TS Values', fontsize=title_font_size, pad=clearance_size)
plt.xticks(fontsize=ticks_font_size)
plt.yticks(fontsize=ticks_font_size)
#plt.grid(True)
#plt.legend(fontsize=legend_font_size)
plt.show()
```

```python
import numpy as np
#import matplotlib.pyplot as plt #標準プロットの場合
import matplotlib.lines as mlines #カスタムプロットの場合
import pickle
import pandas as pd
import os

# 出力フォルダを指定
output_folder = "/content/drive/MyDrive/program/waveform_outputs_DP"
os.makedirs(output_folder, exist_ok=True)

# TS値を定義
ts_values = [500, 538, 546, 600, 607, 700, 770, 800, 907, 1000, 1100, 1190, 1200]

# 保存されたモデルとスケーラーをロード
with open('pca_model.pkl', 'rb') as f:
    pca_model = pickle.load(f)
with open('scaler_ts.pkl', 'rb') as f:
    ts_scaler = pickle.load(f)
with open('scaler_pca.pkl', 'rb') as f:
    pca_scaler = pickle.load(f)
with open('model_pca1_poly.pkl', 'rb') as f:
    model_pca1_poly = pickle.load(f)

# εTS計算関数（指数関数）
def calculate_εTS(ts_value):
    return -0.0002 * ts_value + 0.2987

# 波形生成関数
def generate_waveform_from_real_ts(real_ts_value):
    real_ts_scaled = ts_scaler.transform([[real_ts_value]])[0][0]
    predicted_pca1_scaled = model_pca1_poly.predict([[real_ts_scaled]])
    predicted_pca1 = pca_scaler.inverse_transform(predicted_pca1_scaled.reshape(-1, 1))
    reconstructed_waveform = pca_model.inverse_transform(predicted_pca1)
    return reconstructed_waveform[0]

# 正規化を戻す波形生成関数
def generate_adjusted_waveform(ts_value, target_length=100):
    # 正規化ままの波形を生成
    nominal_strain = np.linspace(0, 1, target_length)
    reconstructed_waveform = generate_waveform_from_real_ts(ts_value)

    # TS時の最大ひずみを取得
    max_nominal_stress = np.max(reconstructed_waveform)
    max_nominal_strain = nominal_strain[np.argmax(reconstructed_waveform)]

    # εTSの計算
    εTS = calculate_εTS(ts_value)

    # 新しい正規化を戻したひずみを計算
    # 波形全体に対する重みを考慮
    adjusted_strain = nominal_strain * (εTS / max_nominal_strain)

    return adjusted_strain, reconstructed_waveform

# 波形を可視化しCSV保存
def plot_and_save_waveforms(ts_values, target_length=100):
    # プロット設定
    title_font_size = 32
    label_font_size = 30
    legend_font_size = 27
    ticks_font_size = 24
    line_width = 2  # カスタム プロットラインの太さ
    legend_line_width = 4  # カスタム 凡例ラインの太さ

    plt.figure(figsize=(10, 8))

    for ts_value in ts_values:
        adjusted_strain, reconstructed_waveform = generate_adjusted_waveform(ts_value, target_length)

        plt.plot(adjusted_strain, reconstructed_waveform, label=f'TS {ts_value}')

        # CSVの保存
        filename = os.path.join(output_folder, f'waveform-DP_TS_{int(ts_value)}.csv')
        df = pd.DataFrame({'Nominal strain': adjusted_strain, 'Nominal stress': reconstructed_waveform})
        df.to_csv(filename, index=False)

    plt.xlabel('Nominal Strain [-]', fontsize=label_font_size,labelpad=20)
    plt.ylabel('Nominal Stress [MPa]', fontsize=label_font_size)
    plt.title('Generative Waveform(Dual-Phase Steels)', fontsize=title_font_size,pad=20)
    plt.legend(fontsize=legend_font_size, loc='upper left', bbox_to_anchor=(1, 1.02))

    # カスタム凡例の作成
    legend_lines = [plt.Line2D([0], [0], color=line.get_color(), label=line.get_label(), linewidth=legend_line_width) for line in plt.gca().get_lines()]
    plt.legend(handles=legend_lines, fontsize=legend_font_size, loc='upper left', bbox_to_anchor=(1, 1.02))  # 凡例を設定

    #標準凡例の場合
    #plt.legend(fontsize=legend_font_size, loc='upper left', bbox_to_anchor=(1, 1))
    #plt.grid(True)

    plt.xticks(fontsize=ticks_font_size)
    plt.yticks(fontsize=ticks_font_size)
    plt.show()

# 実行
plot_and_save_waveforms(ts_values)
```
